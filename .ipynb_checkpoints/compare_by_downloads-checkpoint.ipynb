```python
import matplotlib.pyplot as plt
import json
import re


def parse_downloads(s):
    if not s:
        return 0
    s = str(s).strip().lower()
    if s in ("", "-"):
        return 0
    # handle k, m suffixes and remove commas
    s = s.replace(",", "")
    multiplier = 1
    if s.endswith("k"):
        multiplier = 1_000
        s = s[:-1]
    elif s.endswith("m"):
        multiplier = 1_000_000
        s = s[:-1]
    # extract numeric portion
    m = re.search(r"[\d.]+", s)
    if not m:
        return 0
    try:
        return int(float(m.group(0)) * multiplier)
    except ValueError:
        return 0
```

```python
models_list = []
with open(r"..\ModelsLeaderboard\ModelsLeaderboard\Trending-Models.jsonl", "r", encoding="utf-8") as file:
    for line in file:
        models_list.append(json.loads(line))

models_name = []
total_downloads = []
for data in models_list:
    downloads_raw = data.get("Downloads", "")
    model = data.get("Name", "")
    downloads = parse_downloads(downloads_raw)
    models_name.append(model)
    total_downloads.append(downloads)

plt.figure(figsize=(12, 6))
plt.plot(models_name, total_downloads, marker="o")
plt.xlabel("Models")
plt.ylabel("Downloads")
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```
